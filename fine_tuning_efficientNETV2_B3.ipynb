```python
import tensorflow as tf
import os 
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications import EfficientNetV2B3
from tensorflow.keras import layers
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.efficientnet_v2 import preprocess_input
#300,300,3 for B3
```

```python
#Building model for fine tuning 
def build_model(num_classes):
    IMG_SIZE = 300
    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    model = EfficientNetV2B3(include_top=False, input_tensor=inputs, weights="imagenet")

    # Freeze the pretrained weights
    model.trainable = False

    # Rebuild top
    x = layers.GlobalAveragePooling2D(name="avg_pool")(model.output)
    x = layers.BatchNormalization()(x)

    top_dropout_rate = 0.2
    x = layers.Dropout(top_dropout_rate, name="top_dropout")(x)
    outputs = layers.Dense(num_classes, activation="softmax", name="pred")(x)

    # Compile
    model = tf.keras.Model(inputs, outputs, name="EfficientNet")
    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4)
    model.compile(
        optimizer=optimizer, loss="categorical_crossentropy", metrics=["accuracy"]
    )
    return model

#plotting the learning results 
def plot_hist(hist):
    plt.plot(hist.history["accuracy"])
    plt.plot(hist.history["val_accuracy"])
    plt.title("model accuracy")
    plt.ylabel("accuracy")
    plt.xlabel("epoch")
    plt.legend(["train", "validation"], loc="upper left")
    plt.show()
    
#preproccesing for single file 
def load_and_preprocess_image(image_path, target_size=(300, 300)):
    img = image.load_img(image_path, target_size=target_size)
    img_array = image.img_to_array(img)
    img_array_expanded_dims = np.expand_dims(img_array, axis=0)
    return preprocess_input(img_array_expanded_dims)
```

```python
#Data Preparation 
Num_classes = 4
IMG_SIZE = 300
path = os.getcwd()

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    directory= path + r'\data\train',
    shuffle=True,
    batch_size=5,  
    image_size=(IMG_SIZE, IMG_SIZE),
    label_mode='categorical'
)

test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    directory= path +r'\data\test',
    shuffle=True,
    batch_size=5,
    image_size=(IMG_SIZE, IMG_SIZE),
    label_mode='categorical'
)

#Model fine tuning 
model = build_model(Num_classes)
epochs = 100
hist = model.fit(train_ds, epochs=epochs, validation_data=test_ds)
#Plotting results 
plot_hist(hist)
```

```python
#predicting for new picture
image_path = path + r'\mido.jpg'
img= load_and_preprocess_image(image_path)
class_names = train_ds.class_names
predictions = model.predict(img)
predicted_class_indices = np.argmax(predictions, axis=1)
predicted_class_names = [class_names[i] for i in predicted_class_indices]
print("Predicted class name:", predicted_class_names)
```
